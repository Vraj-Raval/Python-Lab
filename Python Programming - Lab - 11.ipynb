```python
x = range(1,11)
aMarks= [8,9,6,3,5,7,4,1,2,6]
bMarks= [5,8,9,6,3,2,4,8,8,9]

# write a code to generate below graph
plt.plot(x,aMarks,ls="--",marker="*")
plt.plot(x,bMarks,c="b",ls="--",marker="*")
plt.xlabel("Roll No")
plt.ylabel("Cpi")
plt.annotate(xy=[8,1],text="highest Spi")
plt.legend(['cx','cy'],loc=3)
# plt.savefeed()
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
import matplotlib.pyplot as plt
y=['Vraj','Harsh','Ridham']
x=[10,20,30]
plt.bar(y,x,color=['r','g','b'])
plt.title("Bar Chart")
def labels(y,x):
    for i in range(len(y)):
        plt.text(i,x[i],x[i])
labels(y,x)        
plt.show()
```
